```python
import pandas as pd
import zipfile as zp
import os
from sklearn.model_selection import train_test_split

df = pd.read_csv('/kaggle/input/filtered/Filtered (2).csv')
pathology_cols = ['Pneumothorax', 'Cardiomegaly', 'Lung Opacity', 'Pleural Effusion', 'Support Devices']

df['label'] = df[pathology_cols].idxmax(axis=1)
counts = df['label'].value_counts()
df = df[df['label'].isin(counts[counts > 1].index)]

train_val_df, test_df = train_test_split(
    df, test_size=0.15, stratify=df['label'], random_state=42
)

train_val_df.drop(columns=['label']).to_csv('train_val.csv', index=False)
test_df.drop(columns=['label']).to_csv('test.csv', index=False)
```

```python
tv_names = set(train_val_df['Image_name'].astype(str))
t_names = set(test_df['Image_name'].astype(str))

with zp.ZipFile('/kaggle/input/data-seperation/dataset.zip', 'r') as src:
    with zp.ZipFile('train_val_data.zip', 'w') as z_tv, zp.ZipFile('test_data.zip', 'w') as z_t:
        for i, info in enumerate(src.infolist()):
            name = info.filename
            data = src.read(name)
            if name in tv_names:
                z_tv.writestr(name, data)
            elif name in t_names:
                z_t.writestr(name, data)
            if (i + 1) % 1000 == 0:
                print(f"Processed {i + 1} files...")

print("Done!")
```
