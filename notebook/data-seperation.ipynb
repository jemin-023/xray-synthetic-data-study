```python
import os
import pandas as pd
import zipfile as zp

csv_path = '/kaggle/input/filtered/Filtered (2).csv'
source_folder = '/kaggle/input/grand-xray-slam-division-a/train1'
zip_name = '/kaggle/working/dataset.zip'            
image_column_name = 'Image_name' 

df = pd.read_csv(csv_path)
images_to_zip = df[image_column_name].tolist()

print(f"Total images: {len(images_to_zip)}")

with zp.ZipFile(zip_name, 'w', compression=zp.ZIP_DEFLATED) as zf:
    count = 0
    missing = 0
    
    for filename in images_to_zip:
        filename = str(filename)
        source_path = os.path.join(source_folder, filename)
        
        if os.path.exists(source_path):
            zf.write(source_path, arcname=filename)
            count += 1
            if count % 500 == 0:
                print(f"Archived {count} images")
        else:
            missing += 1

print("Done")
```
